```python
import json
from pathlib import Path
import pandas as pd
import duckdb

def _top_key(d):
    if isinstance(d, dict) and d:
        return max(d, key=d.get)  # key with highest value
    return None

def load_smogon_usage(
    json_path: str,
    db_path: str = "poke.duckdb",
    table_name: str = "smogon_usage"
):
    """
    Load a Smogon usage JSON file, flatten key fields, and store as a DuckDB table.
    Extracts: name, usage, raw_count, top_ability, top_item, top_move.
    Returns the prepared pandas DataFrame.
    """
    p = Path(json_path)
    if not p.exists():
        raise FileNotFoundError(f"JSON not found at {p}")

    with p.open("r") as f:
        raw = json.load(f)

    # Most Smogon dumps have top-level "data" keyed by Pokemon name
    container = raw.get("data", raw)

    rows = []
    for name, info in container.items():
        # Usage & raw count fields vary by dump â€“ try common variants
        usage = (
            info.get("usage")
            or info.get("Usage")
            or info.get("usage %")
            or info.get("Usage %")
        )

        raw_count = (
            info.get("Raw count")
            or info.get("Raw Count")
            or info.get("raw_count")
            or info.get("Raw counts")
        )
        if isinstance(raw_count, dict):
            # sometimes a dict; try to pull a sensible number
            raw_count = raw_count.get("usage") or raw_count.get("Usage") or raw_count.get("raw") or None

        abilities = info.get("Abilities") or info.get("abilities")
        items     = info.get("Items") or info.get("items")
        moves     = info.get("Moves") or info.get("moves")

        rows.append({
            "name": name,
            "usage": usage,
            "raw_count": raw_count,
            "top_ability": _top_key(abilities),
            "top_item": _top_key(items),
            "top_move": _top_key(moves),
        })

    df = pd.DataFrame(rows)

    # Persist to DuckDB
    con = duckdb.connect(db_path)
    con.register("smogon_df", df)
    con.execute(f"""
        CREATE OR REPLACE TABLE {table_name} AS
        SELECT * FROM smogon_df
    """)
    con.close()
    return df
```

```python
import json
from pathlib import Path
import pandas as pd
import duckdb

def _top_key(d):
    """Return key with highest value from a dict, else None."""
    if isinstance(d, dict) and d:
        return max(d, key=d.get)
    return None

def _top_n_keys(d, n=4):
    """Return list of top-n keys by value from a dict (length may be < n)."""
    if not isinstance(d, dict) or not d:
        return []
    return [k for k, _ in sorted(d.items(), key=lambda x: x[1], reverse=True)[:n]]

def _parse_nature_from_spread(spread_key):
    """
    Smogon 'Spreads' keys usually look like: 'Timid:60/0/0/196/0/252'
    Return the part before the colon as the nature.
    """
    if not isinstance(spread_key, str):
        return None
    return spread_key.split(':', 1)[0] if ':' in spread_key else None

def load_smogon_usage(
    json_path: str,
    db_path: str = "poke.duckdb",
    table_name: str = "smogon_usage"
):
    """
    Load a Smogon usage JSON file, flatten key fields, and store as a DuckDB table.

    Extracted columns:
      - name
      - usage
      - raw_count
      - top_ability
      - top_item
      - top_tera_type
      - top_spread
      - top_nature
      - move1, move2, move3, move4 (four most-used moves)
    """
    p = Path(json_path)
    if not p.exists():
        raise FileNotFoundError(f"JSON not found at {p}")

    with p.open("r") as f:
        raw = json.load(f)

    container = raw.get("data", raw)  # many dumps store the data under 'data'

    rows = []
    for name, info in container.items():
        # Usage and raw count can be named differently across dumps
        usage = (
            info.get("usage")
            or info.get("Usage")
            or info.get("usage %")
            or info.get("Usage %")
        )

        raw_count = (
            info.get("Raw count")
            or info.get("Raw Count")
            or info.get("raw_count")
            or info.get("Raw counts")
        )
        if isinstance(raw_count, dict):
            raw_count = (
                raw_count.get("usage") or raw_count.get("Usage")
                or raw_count.get("raw") or None
            )

        abilities  = info.get("Abilities")  or info.get("abilities")
        items      = info.get("Items")      or info.get("items")
        moves      = info.get("Moves")      or info.get("moves")
        spreads    = info.get("Spreads")    or info.get("spreads")
        tera_types = info.get("Tera Types") or info.get("Tera types") or info.get("tera_types")

        # Top picks
        top_ability   = _top_key(abilities)
        top_item      = _top_key(items)
        top_tera_type = _top_key(tera_types)
        top_spread    = _top_key(spreads)
        top_nature    = _parse_nature_from_spread(top_spread)

        # Four most used moves
        top_moves = _top_n_keys(moves, n=4)
        # Pad to length 4 so columns are consistent
        while len(top_moves) < 4:
            top_moves.append(None)

        rows.append({
            "name": name,
            "usage": usage,
            "raw_count": raw_count,
            "top_ability": top_ability,
            "top_item": top_item,
            "top_tera_type": top_tera_type,
            "top_spread": top_spread,
            "top_nature": top_nature,
            "move1": top_moves[0],
            "move2": top_moves[1],
            "move3": top_moves[2],
            "move4": top_moves[3],
        })

    df = pd.DataFrame(rows)

    # Persist to DuckDB
    con = duckdb.connect(db_path)
    con.register("smogon_df", df)
    con.execute(f"""
        CREATE OR REPLACE TABLE {table_name} AS
        SELECT * FROM smogon_df
    """)
    con.close()

    return df
```

```python
import json
from pathlib import Path
import pandas as pd
import duckdb

def _top_with_pct(d):
    """Return (key, value) with the highest value from a dict, else (None, None)."""
    if isinstance(d, dict) and d:
        k = max(d, key=d.get)
        return k, d[k]
    return None, None

def _top_n_with_pct(d, n=4):
    """Return list of (key, value) pairs for the top-n entries by value."""
    if not isinstance(d, dict) or not d:
        return []
    return sorted(d.items(), key=lambda x: x[1], reverse=True)[:n]

def _parse_nature_from_spread(spread_key):
    # e.g. "Timid:60/0/0/196/0/252" -> "Timid"
    if not isinstance(spread_key, str):
        return None
    return spread_key.split(':', 1)[0] if ':' in spread_key else None

def _strip_nature_from_spread(spread_key):
    # e.g. "Timid:60/0/0/196/0/252" -> "60/0/0/196/0/252"
    if not isinstance(spread_key, str):
        return None
    return spread_key.split(':', 1)[1] if ':' in spread_key else spread_key

def load_smogon_usage(
    json_path: str,
    db_path: str = "poke.duckdb",
    table_name: str = "smogon_usage"
):
    p = Path(json_path)
    with p.open("r") as f:
        raw = json.load(f)

    container = raw.get("data", raw)

    rows = []
    for name, info in container.items():
        usage = (info.get("usage") or info.get("Usage") or
                 info.get("usage %") or info.get("Usage %"))

        raw_count = (info.get("Raw count") or info.get("Raw Count") or
                     info.get("raw_count") or info.get("Raw counts"))
        if isinstance(raw_count, dict):
            raw_count = (raw_count.get("usage") or raw_count.get("Usage") or
                         raw_count.get("raw"))

        abilities  = info.get("Abilities")  or info.get("abilities")
        items      = info.get("Items")      or info.get("items")
        moves      = info.get("Moves")      or info.get("moves")
        spreads    = info.get("Spreads")    or info.get("spreads")
        tera_types = (info.get("Tera Types") or info.get("Tera types") or
                      info.get("tera_types"))

        top_ability, top_ability_pct = _top_with_pct(abilities)
        top_item,    top_item_pct    = _top_with_pct(items)
        top_tera,    top_tera_pct    = _top_with_pct(tera_types)
        top_spread,  top_spread_pct  = _top_with_pct(spreads)  # keep pct too if you want

        # four most-used moves with their values
        top_moves = _top_n_with_pct(moves, 4)
        # pad to always have 4
        top_moves += [(None, None)] * (4 - len(top_moves))

        rows.append({
            "name": name,
            "usage": usage,
            "raw_count": raw_count,
            "top_ability": top_ability,
            "top_item": top_item,
            "top_item_pct": top_item_pct,
            "top_tera_type": top_tera,
            "top_tera_pct": top_tera_pct,
            "top_spread": top_spread,
            "top_spread_no_nature": _strip_nature_from_spread(top_spread),
            "top_nature": _parse_nature_from_spread(top_spread),
            "move1": top_moves[0][0], "move1_pct": top_moves[0][1],
            "move2": top_moves[1][0], "move2_pct": top_moves[1][1],
            "move3": top_moves[2][0], "move3_pct": top_moves[2][1],
            "move4": top_moves[3][0], "move4_pct": top_moves[3][1],
        })

    df = pd.DataFrame(rows)

    con = duckdb.connect(db_path)
    con.register("smogon_df", df)
    con.execute(f"CREATE OR REPLACE TABLE {table_name} AS SELECT * FROM smogon_df")
    con.close()
    return df
```

```python
import json
from pathlib import Path
import pandas as pd
import duckdb

def _normalize(d, n=1):
    """Return a dict with values normalized so they sum to n (default = 1.0)."""
    if not isinstance(d, dict) or not d:
        return {}
    total = sum(d.values())
    if total == 0:
        return {}
    return {k: (v / total) * n for k, v in d.items()}

def _top_with_pct(d):
    """Return (key, pct) for the highest entry in a normalized dict."""
    if not d:
        return None, None
    k = max(d, key=d.get)
    return k, d[k]

def _top_n_with_pct(d, n=4):
    """Return top-n (key, pct) from normalized dict."""
    if not d:
        return []
    return sorted(d.items(), key=lambda x: x[1], reverse=True)[:n]

def _parse_nature_from_spread(spread_key):
    if not isinstance(spread_key, str):
        return None
    return spread_key.split(':', 1)[0] if ':' in spread_key else None

def _strip_nature_from_spread(spread_key):
    if not isinstance(spread_key, str):
        return None
    return spread_key.split(':', 1)[1] if ':' in spread_key else spread_key

def load_smogon_usage(
    json_path: str,
    db_path: str = "poke.duckdb",
    table_name: str = "smogon_usage"
):
    p = Path(json_path)
    with p.open("r") as f:
        raw = json.load(f)

    container = raw.get("data", raw)

    rows = []
    for name, info in container.items():
        usage = info.get("usage") or info.get("Usage") or None
        raw_count = info.get("Raw count") or info.get("raw_count")

        # Normalize dicts
        abilities  = _normalize(info.get("Abilities")  or {})
        items      = _normalize(info.get("Items")      or {})
        moves      = _normalize(info.get("Moves")      or {})
        spreads    = _normalize(info.get("Spreads")    or {})
        tera_types = _normalize(info.get("Tera Types") or {})

        top_ability, top_ability_pct = _top_with_pct(abilities)
        top_item,    top_item_pct    = _top_with_pct(items)
        top_tera,    top_tera_pct    = _top_with_pct(tera_types)
        top_spread,  top_spread_pct  = _top_with_pct(spreads)

        top_moves = _top_n_with_pct(moves, 4)
        top_moves += [(None, None)] * (4 - len(top_moves))

        rows.append({
            "name": name,
            "usage": usage,
            "raw_count": raw_count,
            "top_ability": top_ability,
            "top_item": top_item, "top_item_pct": top_item_pct,
            "top_tera_type": top_tera, "top_tera_pct": top_tera_pct,
            "top_spread": top_spread,
            "top_spread_no_nature": _strip_nature_from_spread(top_spread),
            "top_nature": _parse_nature_from_spread(top_spread),
            "move1": top_moves[0][0], "move1_pct": top_moves[0][1],
            "move2": top_moves[1][0], "move2_pct": top_moves[1][1],
            "move3": top_moves[2][0], "move3_pct": top_moves[2][1],
            "move4": top_moves[3][0], "move4_pct": top_moves[3][1],
        })

    df = pd.DataFrame(rows)

    # Save to DuckDB
    con = duckdb.connect(db_path)
    con.register("smogon_df", df)
    con.execute(f"CREATE OR REPLACE TABLE {table_name} AS SELECT * FROM smogon_df")
    con.close()

    return df
```

```python
import json
from pathlib import Path
import pandas as pd
import duckdb

def _normalize_single_choice(d):
    """Normalize a dict that represents one choice per set (items, abilities, spreads, tera)."""
    if not isinstance(d, dict) or not d:
        return {}
    tot = sum(d.values())
    if tot == 0:
        return {}
    return {k: v / tot for k, v in d.items()}  # sums to 1.0 across choices

def _moves_absolute_pct(d):
    """
    Convert moves dict to absolute per-set percentages.
    Each set has 4 moves, so set_count â‰ˆ sum(values)/4.
    Return dict where values are in [0,1] and indicate
    'share of sets that include this move'.
    """
    if not isinstance(d, dict) or not d:
        return {}
    total = sum(d.values())
    if total == 0:
        return {}
    set_count = total / 4.0
    if set_count == 0:
        return {}
    return {k: (v / set_count) for k, v in d.items()}  # = v * 4 / total

def _top_with_pct(d):
    if not d:
        return None, None
    k = max(d, key=d.get)
    return k, d[k]

def _top_n_with_pct(d, n=4):
    if not d:
        return []
    return sorted(d.items(), key=lambda x: x[1], reverse=True)[:n]

def _parse_nature_from_spread(spread_key):
    if not isinstance(spread_key, str):
        return None
    return spread_key.split(':', 1)[0] if ':' in spread_key else None

def _strip_nature_from_spread(spread_key):
    if not isinstance(spread_key, str):
        return None
    return spread_key.split(':', 1)[1] if ':' in spread_key else spread_key

def load_smogon_usage(json_path, db_path="poke.duckdb", table_name="smogon_usage"):
    p = Path(json_path)
    with p.open("r") as f:
        raw = json.load(f)

    container = raw.get("data", raw)

    rows = []
    for name, info in container.items():
        usage = info.get("usage") or info.get("Usage") or None
        raw_count = info.get("Raw count") or info.get("raw_count")

        # single-choice per set â†’ normalize to 0..1
        abilities  = _normalize_single_choice(info.get("Abilities")  or {})
        items      = _normalize_single_choice(info.get("Items")      or {})
        spreads    = _normalize_single_choice(info.get("Spreads")    or {})
        tera_types = _normalize_single_choice(info.get("Tera Types") or {})

        # moves â†’ absolute % of sets
        moves_abs  = _moves_absolute_pct(info.get("Moves") or {})

        top_ability, top_ability_pct = _top_with_pct(abilities)
        top_item,    top_item_pct    = _top_with_pct(items)
        top_tera,    top_tera_pct    = _top_with_pct(tera_types)
        top_spread,  top_spread_pct  = _top_with_pct(spreads)

        top_moves = _top_n_with_pct(moves_abs, 4)
        top_moves += [(None, None)] * (4 - len(top_moves))

        rows.append({
            "name": name,
            "usage": usage,
            "raw_count": raw_count,
            "top_ability": top_ability,
            "top_item": top_item, "top_item_pct": top_item_pct,
            "top_tera_type": top_tera, "top_tera_pct": top_tera_pct,
            "top_spread": top_spread,
            "top_spread_no_nature": _strip_nature_from_spread(top_spread),
            "top_nature": _parse_nature_from_spread(top_spread),
            "move1": top_moves[0][0], "move1_pct": top_moves[0][1],
            "move2": top_moves[1][0], "move2_pct": top_moves[1][1],
            "move3": top_moves[2][0], "move3_pct": top_moves[2][1],
            "move4": top_moves[3][0], "move4_pct": top_moves[3][1],
        })

    df = pd.DataFrame(rows)

    con = duckdb.connect(db_path)
    con.register("smogon_df", df)
    con.execute(f"CREATE OR REPLACE TABLE {table_name} AS SELECT * FROM smogon_df")
    con.close()
    return df
```

```python
def show_profile(pokemon: str):
    """
    Render a one-page profile from smogon_usage:
    - usage %
    - top ability
    - top item (+ %)
    - top Tera type (+ %)
    - top spread (no nature) + top nature
    - top 4 moves (+ % each)
    Also draws a small bar chart for the top-4 moves.
    """
      name = _resolve_name_interactive(pokemon)
    if not name:  # user cancelled or no match
        return
        raise ValueError(f"Couldn't find a close match for '{pokemon}' in smogon_usage")

    q = """
    SELECT
      name,
      ROUND(usage * 100, 2)        AS usage_pct,
      top_ability,
      top_item,  ROUND(top_item_pct * 100, 2)  AS item_pct,
      top_tera_type, ROUND(top_tera_pct * 100, 2) AS tera_pct,
      top_spread_no_nature, top_nature,
      move1, ROUND(move1_pct * 100, 2) AS move1_pct,
      move2, ROUND(move2_pct * 100, 2) AS move2_pct,
      move3, ROUND(move3_pct * 100, 2) AS move3_pct,
      move4, ROUND(move4_pct * 100, 2) AS move4_pct
    FROM smogon_usage
    WHERE lower(name) = lower(?)
    """
    df = con.execute(q, [name]).fetchdf()
    if df.empty:
        raise ValueError(f"No row found for '{name}'")

    display_df = df[[
        "name", "usage_pct",
        "top_ability",
        "top_item", "item_pct",
        "top_tera_type", "tera_pct",
        "top_spread_no_nature", "top_nature",
        "move1","move1_pct","move2","move2_pct","move3","move3_pct","move4","move4_pct"
    ]].copy()

    # Show the profile table
    display_df.index = ["Profile"]
    display(display_df)

    # Build a moves dataframe for the chart
    moves = pd.DataFrame({
        "move": [df.at[0,"move1"], df.at[0,"move2"], df.at[0,"move3"], df.at[0,"move4"]],
        "pct":  [df.at[0,"move1_pct"], df.at[0,"move2_pct"], df.at[0,"move3_pct"], df.at[0,"move4_pct"]],
    }).dropna()

    # Bar chart for top-4 moves (matplotlib, single plot, default colors)
    plt.figure(figsize=(8,4))
    plt.bar(moves["move"], moves["pct"])
    plt.title(f"{name}: Top-4 Move Usage (% of sets)")
    plt.ylabel("% of sets")
    plt.xlabel("Move")
    plt.xticks(rotation=15)
    plt.tight_layout()
    plt.show()

    # Small text summary (nice to copy into notes)
    summary = (
        f"{name} â€” Usage: {df.at[0,'usage_pct']}%\n"
        f"Ability: {df.at[0,'top_ability']} | Item: {df.at[0,'top_item']} ({df.at[0,'item_pct']}%) | "
        f"Tera: {df.at[0,'top_tera_type']} ({df.at[0,'tera_pct']}%)\n"
        f"Spread: {df.at[0,'top_nature']} {df.at[0,'top_spread_no_nature']}\n"
        f"Moves: {df.at[0,'move1']} ({df.at[0,'move1_pct']}%), "
        f"{df.at[0,'move2']} ({df.at[0,'move2_pct']}%), "
        f"{df.at[0,'move3']} ({df.at[0,'move3_pct']}%), "
        f"{df.at[0,'move4']} ({df.at[0,'move4_pct']}%)"
    )
    print(summary)
```

```python
def _get_exact_or_close_name(query: str) -> str:
    names = con.execute("SELECT DISTINCT name FROM smogon_usage").fetchdf()["name"].tolist()

    # Exact match (case-insensitive)
    for n in names:
        if n.lower() == query.lower():
            return n

    # Substring match
    subs = [n for n in names if query.lower() in n.lower()]
    if subs:
        # If multiple, pick the shortest (so "zacian" â†’ "Zacian-Crowned")
        return min(subs, key=len)

    # Fuzzy match fallback
    from difflib import get_close_matches
    hits = get_close_matches(query, names, n=1, cutoff=0.5)
    return hits[0] if hits else None
```

```python
from difflib import get_close_matches

def _resolve_name_interactive(query: str) -> str | None:
    """Resolve a PokÃ©mon name from a partial string.
    - exact case-insensitive match wins
    - otherwise substring matches
    - otherwise fuzzy matches (top 5)
    If multiple candidates, show a numbered list and ask you to pick.
    Returns the chosen name or None if cancelled/not found.
    """
    names = con.execute("SELECT DISTINCT name FROM smogon_usage").fetchdf()["name"].tolist()
    q = query.strip().lower()

    # Exact match
    for n in names:
        if n.lower() == q:
            return n

    # Substring matches
    subs = [n for n in names if q in n.lower()]

    candidates = subs
    if not candidates:
        # Fuzzy (top 5) if no substring hits
        candidates = get_close_matches(query, names, n=5, cutoff=0.45)

    if not candidates:
        print(f"No match found for '{query}'.")
        return None

    if len(candidates) == 1:
        return candidates[0]

    # Ask the user to choose
    print(f"Multiple matches for '{query}':")
    for i, n in enumerate(candidates, 1):
        print(f"  {i}. {n}")
    while True:
        try:
            sel = int(input(f"Choose 1â€“{len(candidates)} (0 to cancel): ").strip())
            if sel == 0:
                return None
            if 1 <= sel <= len(candidates):
                return candidates[sel - 1]
        except ValueError:
            pass
        print("Invalid choice, try again.")
```

```python
import duckdb, pandas as pd, matplotlib.pyplot as plt, ipywidgets as widgets
from IPython.display import display, clear_output
from difflib import get_close_matches

con = duckdb.connect("poke.duckdb")
plt.rcParams["figure.figsize"] = (8,4)
```

```python
def show_profile_exact(name: str):
    q = """
    SELECT
      name,
      ROUND(usage * 100, 2) AS usage_pct,
      top_ability,
      top_item,       ROUND(top_item_pct * 100, 2)  AS item_pct,
      top_tera_type,  ROUND(top_tera_pct * 100, 2)  AS tera_pct,
      top_spread_no_nature, top_nature,
      move1, ROUND(move1_pct * 100, 2) AS move1_pct,
      move2, ROUND(move2_pct * 100, 2) AS move2_pct,
      move3, ROUND(move3_pct * 100, 2) AS move3_pct,
      move4, ROUND(move4_pct * 100, 2) AS move4_pct
    FROM smogon_usage
    WHERE name = ?
    """
    df = con.execute(q, [name]).fetchdf()
    if df.empty:
        print(f"No row found for '{name}'"); return

    prof = df[["name","usage_pct","top_ability",
               "top_item","item_pct","top_tera_type","tera_pct",
               "top_spread_no_nature","top_nature",
               "move1","move1_pct","move2","move2_pct","move3","move3_pct","move4","move4_pct"]]
    prof.index = ["Profile"]
    display(prof)

    # bar chart for top-4 moves
    moves = pd.DataFrame({
        "move": [df.at[0,"move1"], df.at[0,"move2"], df.at[0,"move3"], df.at[0,"move4"]],
        "pct":  [df.at[0,"move1_pct"], df.at[0,"move2_pct"], df.at[0,"move3_pct"], df.at[0,"move4_pct"]],
    }).dropna()
    plt.figure()
    plt.bar(moves["move"], moves["pct"])
    plt.title(f"{name}: Top-4 Move Usage (% of sets)")
    plt.ylabel("% of sets"); plt.xlabel("Move"); plt.xticks(rotation=15); plt.tight_layout(); plt.show()
```

```python
_all_names = con.execute("SELECT DISTINCT name FROM smogon_usage").fetchdf()["name"].tolist()

def find_candidates(query: str, max_results: int = 10):
    q = (query or "").strip().lower()
    if not q:
        return []
    subs = [n for n in _all_names if q in n.lower()]
    if subs:
        return sorted(subs, key=len)[:max_results]
    return get_close_matches(query, _all_names, n=max_results, cutoff=0.45)
```

```python
# ---------- Widgets ----------
top_n     = widgets.IntSlider(value=20, min=5, max=50, step=5, description="Top N:")
refresh   = widgets.Button(description="Refresh Top")
out_top   = widgets.Output()

query_box = widgets.Text(placeholder="type a PokÃ©mon (e.g., zacian)", description="Search:")
find_btn  = widgets.Button(description="Find matches")
pick_dd   = widgets.Dropdown(options=[], description="Matches:", disabled=True)
show_btn  = widgets.Button(description="Show profile", disabled=True)
out_prof  = widgets.Output()

out_items = widgets.Output()
out_tera  = widgets.Output()

# ---------- Handlers ----------
def render_top():
    with out_top:
        clear_output()
        df = con.execute(f"""
            SELECT
              name,
              ROUND(usage * 100, 2) AS usage_pct,
              top_item, ROUND(top_item_pct * 100, 2) AS item_pct,
              top_tera_type, ROUND(top_tera_pct * 100, 2) AS tera_pct
            FROM smogon_usage
            WHERE usage IS NOT NULL
            ORDER BY usage DESC
            LIMIT {int(top_n.value)}
        """).fetchdf()
        display(df)

def on_refresh_clicked(_):
    render_top()

def on_find_clicked(_):
    with out_prof:
        clear_output()
    cands = find_candidates(query_box.value)
    if not cands:
        pick_dd.options = []; pick_dd.disabled = True; show_btn.disabled = True
        with out_prof:
            print(f"No matches for '{query_box.value}'.")
        return
    pick_dd.options = cands; pick_dd.value = cands[0]
    pick_dd.disabled = False; show_btn.disabled = False

def on_show_clicked(_):
    with out_prof:
        clear_output()
        show_profile_exact(pick_dd.value)

def render_items_tera():
    # Items distribution (top 12)
    with out_items:
        clear_output()
        df_i = con.execute("""
            SELECT top_item AS item, COUNT(*) AS mons
            FROM smogon_usage
            WHERE top_item IS NOT NULL
            GROUP BY top_item
            ORDER BY mons DESC
            LIMIT 12
        """).fetchdf()
        plt.figure()
        plt.bar(df_i["item"], df_i["mons"])
        plt.title("Most common TOP item across the meta (count of PokÃ©mon)")
        plt.ylabel("PokÃ©mon count"); plt.xlabel("Item")
        plt.xticks(rotation=25); plt.tight_layout(); plt.show()

    # Tera distribution (top 12)
    with out_tera:
        clear_output()
        df_t = con.execute("""
            SELECT top_tera_type AS tera, COUNT(*) AS mons
            FROM smogon_usage
            WHERE top_tera_type IS NOT NULL
            GROUP BY top_tera_type
            ORDER BY mons DESC
            LIMIT 12
        """).fetchdf()
        plt.figure()
        plt.bar(df_t["tera"], df_t["mons"])
        plt.title("Most common TOP Tera type across the meta (count of PokÃ©mon)")
        plt.ylabel("PokÃ©mon count"); plt.xlabel("Tera Type")
        plt.xticks(rotation=25); plt.tight_layout(); plt.show()

# Wire handlers
refresh.on_click(on_refresh_clicked)
find_btn.on_click(on_find_clicked)
show_btn.on_click(on_show_clicked)

# Initial render
render_top(); render_items_tera()

# ---------- Layout ----------
left  = widgets.VBox([widgets.HBox([top_n, refresh]), out_top])
right = widgets.VBox([query_box, widgets.HBox([find_btn, pick_dd, show_btn]), out_prof,
                      widgets.Label("Distributions"), out_items, out_tera])
display(widgets.HBox([left, widgets.HTML("&nbsp;&nbsp;&nbsp;"), right]))
```

```python
import duckdb

con = duckdb.connect("poke.duckdb")

# See current schema
print(con.execute("PRAGMA table_info('smogon_usage')").fetchdf())

# If snapshot_month is not TEXT/VARCHAR, convert it
try:
    con.execute("ALTER TABLE smogon_usage ALTER COLUMN snapshot_month TYPE TEXT")
    print("âœ… snapshot_month changed to TEXT")
except Exception as e:
    # Some older DuckDB builds don't support ALTER TYPE; recreate table safely
    print("ALTER TYPE failed, recreating tableâ€¦", e)
    con.execute("""
        CREATE TABLE smogon_usage__new AS
        SELECT
          CAST(name AS TEXT) AS name,
          CAST(snapshot_month AS TEXT) AS snapshot_month,
          CAST(usage AS DOUBLE) AS usage,
          CAST(raw_count AS DOUBLE) AS raw_count,
          CAST(top_ability AS TEXT) AS top_ability,
          CAST(top_item AS TEXT) AS top_item,
          CAST(top_item_pct AS DOUBLE) AS top_item_pct,
          CAST(top_tera_type AS TEXT) AS top_tera_type,
          CAST(top_tera_pct AS DOUBLE) AS top_tera_pct,
          CAST(top_spread AS TEXT) AS top_spread,
          CAST(top_spread_no_nature AS TEXT) AS top_spread_no_nature,
          CAST(top_nature AS TEXT) AS top_nature,
          CAST(move1 AS TEXT) AS move1, CAST(move1_pct AS DOUBLE) AS move1_pct,
          CAST(move2 AS TEXT) AS move2, CAST(move2_pct AS DOUBLE) AS move2_pct,
          CAST(move3 AS TEXT) AS move3, CAST(move3_pct AS DOUBLE) AS move3_pct,
          CAST(move4 AS TEXT) AS move4, CAST(move4_pct AS DOUBLE) AS move4_pct
        FROM smogon_usage
    """)
    con.execute("DROP TABLE smogon_usage")
    con.execute("ALTER TABLE smogon_usage__new RENAME TO smogon_usage")
    print("âœ… recreated smogon_usage with snapshot_month TEXT")

print(con.execute("PRAGMA table_info('smogon_usage')").fetchdf())
con.close()
```
